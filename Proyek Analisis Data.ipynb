### Explore main_df

```python
# Mengonversi kolom 'dteday' menjadi format datetime
day_df['dteday'] = pd.to_datetime(day_df['dteday'])
hour_df['dteday'] = pd.to_datetime(hour_df['dteday'])

# Menggabungkan day_df dan hour_df
main_df = pd.merge(
    left=day_df,            # DataFrame yang akan menjadi 'kiri' dalam penggabungan
    right=hour_df,          # DataFrame yang akan menjadi 'kanan' dalam penggabungan
    how="left",             # Jenis penggabungan yang dilakukan (dalam hal ini, "left join")
    left_on="dteday",       # Kolom kunci yang akan digunakan untuk DataFrame 'kiri' (day_df)
    right_on="dteday"       # Kolom kunci yang akan digunakan untuk DataFrame 'kanan' (hour_df)
)

# Memeriksa 5 baris pertama dari DataFrame hasil penggabungan
main_df.head()
```

```python
# Table Preview

main_df.info()
```

## Visualization & Explanatory Analysis
### Visualize the overall bike rental trends

```python
# Combining data on a monthly basis

monthly_df = main_df.resample('M', on='dteday').sum()

# Plotting the data

plt.figure(figsize=(10, 6))
plt.plot(monthly_df.index, monthly_df['cnt_x'], color='#6499E9')
plt.xlabel('Month')
plt.ylabel('Number of Bike Rentals')
plt.title('Bike Rental Trends Over Time (Grouped by Month)')
plt.xticks(rotation=45)
plt.grid(True)
plt.tight_layout()
plt.show()
```

### Visualize bike rentals based on season.

```python
# Grouping data based on seasons
season_agg = main_df.groupby("season_x").agg({
    "instant_x": "nunique",
    "cnt_x": ["max", "min"]
})

# Plotting the data
plt.figure(figsize=(10, 6))
x = season_agg.index
y_max = season_agg[('cnt_x', 'max')]
y_min = season_agg[('cnt_x', 'min')]

plt.bar(x, y_max, label='Max Rentals', color='#6499E9')
plt.bar(x, y_min, label='Min Rentals', color='orange')

# Adjusting x-axis labels
season_labels = ['Season 1', 'Season 2', 'Season 3', 'Season 4']
plt.xticks(x, season_labels)

plt.xlabel('Season')
plt.ylabel('Number of Bike Rentals')
plt.title('Max and Min Bike Rentals by Season')
plt.legend()

# Annotating maximum and minimum values
for i, (max_val, min_val) in enumerate(zip(y_max, y_min)):
    plt.text(i+1, max_val, str(max_val), ha='center', va='bottom', fontweight='bold')
    plt.text(i+1, min_val, str(min_val), ha='center', va='bottom', fontweight='bold')

plt.grid(True)
plt.tight_layout()
plt.show()
```

### Visualize bike rentals based on month.

```python
# Aggregate the data by month
monthly_agg = main_df.groupby("mnth_x").agg({
    "instant_x": "nunique",
    "cnt_x": ["max", "min"]
})

# Plot the data
plt.figure(figsize=(10, 6))
x = monthly_agg.index
y_max = monthly_agg[('cnt_x', 'max')]
y_min = monthly_agg[('cnt_x', 'min')]

plt.bar(x, y_max, label='Max Rentals', color='#6499E9')
plt.bar(x, y_min, label='Min Rentals', color='orange')

# Adjust y-axis labels
month_labels = ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec']
plt.xticks(x, month_labels)

plt.xlabel('Month')
plt.ylabel('Number of Bike Rentals')
plt.title('Max and Min Bike Rentals per Month')
plt.legend()

# Annotate the maximum and minimum values
for i, (max_val, min_val) in enumerate(zip(y_max, y_min)):
    plt.text(i+1, max_val, str(max_val), ha='center', va='bottom', fontweight='bold')
    plt.text(i+1, min_val, str(min_val), ha='center', va='bottom', fontweight='bold')

plt.grid(True)
plt.tight_layout()
plt.show()
```

### Visualize bike rentals based on day.

```python
# Grouping data by weekday
weekday_agg = main_df.groupby("weekday_x").agg({
    "instant_x": "nunique",
    "cnt_x": ["max", "min"]
})

# Plotting the data
plt.figure(figsize=(10, 6))
x = weekday_agg.index
y_max = weekday_agg[('cnt_x', 'max')]
y_min = weekday_agg[('cnt_x', 'min')]

plt.bar(x, y_max, label='Max Rentals', color='#6499E9')
plt.bar(x, y_min, label='Min Rentals', color='orange')

plt.xlabel('Day')
plt.ylabel('Number of Bike Rentals')
plt.title('Max and Min Bike Rentals by Day')
plt.xticks(rotation=0)
plt.legend()

# Annotating maximum and minimum values
for i, (max_val, min_val) in enumerate(zip(y_max, y_min)):
    plt.text(i, max_val, str(max_val), ha='center', va='bottom', fontweight='bold')
    plt.text(i, min_val, str(min_val), ha='center', va='bottom', fontweight='bold')

# Adjusting the legend's position
plt.legend(loc='upper right', bbox_to_anchor=(1.15, 1))

plt.grid(True)
plt.tight_layout()
plt.show()
```
